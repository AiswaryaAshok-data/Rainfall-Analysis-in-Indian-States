```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#We convert the datasets obtained from WRIS India, containing average monthly rainfall in the relevant Indian states, to dataframes

mp = pd.read_csv("mp.csv", header=1, usecols=["Dates", "ACTUAL (mm) "])
chat = pd.read_csv("chat.csv", header=1, usecols=["ACTUAL (mm) "])
guj = pd.read_csv("guj.csv", header=1, usecols=["ACTUAL (mm) "])
mh = pd.read_csv("mh.csv", header=1, usecols=["ACTUAL (mm) "])
raja = pd.read_csv("raja.csv", header=1, usecols=["ACTUAL (mm) "])
up = pd.read_csv("up.csv", header=1, usecols=["ACTUAL (mm) "])
assam = pd.read_csv("assam.csv", header=1, usecols=["ACTUAL (mm) "])
jk = pd.read_csv("jk.csv", header=1, usecols=["ACTUAL (mm) "])
kl = pd.read_csv("kl.csv", header=1, usecols=["ACTUAL (mm) "])
tn = pd.read_csv("tn.csv", header=1, usecols=["ACTUAL (mm) "])

#We check for any missing values
print(mp.isnull().values.any(), chat.isnull().values.any(), guj.isnull().values.any(), mh.isnull().values.any(), raja.isnull().values.any(), up.isnull().values.any(), assam.isnull().values.any(), jk.isnull().values.any(), kl.isnull().values.any(), tn.isnull().values.any(), "\n\n")


#Now, we merge the dataframes and rename the columns suitably
data = pd.concat([mp, chat, guj, mh, raja, up, assam, jk, kl, tn], axis=1)
data = data.set_axis(["Month", "mp", "chat", "guj", "mh", "raja", "up", "assam", "jk", "kl", "tn"], axis = 1)
# print(data.head())
```

```python
"""
Now, we move to the prediction part. We wish to predict the daily rainfall in Madhya Pradesh using rainfall data from other states.
We achieve this by using three sets of input data, one containing rainfal
of neighbout states, one containing rainfall in distant states,
and another containing the rainfall data of all 9 states.

As predicted from the correlation matrix, model with the neighbouring states as input is expected to perform on par with the one containing the entire dataset.
"""


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

x = data_daily.drop(["Dates", "mp"], axis=1)
y = data_daily["mp"]                #For all states' data

print(x.head())

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)           #Scaling the training set and test set


rf = RandomForestRegressor()
gbr = GradientBoostingRegressor()
knr = KNeighborsRegressor()
lr = LinearRegression()         #Declaring base regressors for the ensemble based VotingRegressor

ens = VotingRegressor(estimators=[
    ('rf', rf),
    ('gbr', gbr),
    ('knr', knr),
    ('lr', lr)], verbose=False)


params = {          
    'rf__criterion': ('squared_error', 'absolute_error', 'friedman_mse', 'poisson'),            #Parameters for GridsearchCV
    'gbr__loss': ('squared_error', 'absolute_error', 'huber', 'quantile'),
    'gbr__criterion': ('friedman_mse', 'squared_error'),
    'knr__n_neighbors': (3, 5, 7, 9),
    'knr__algorithm': ('auto', 'ball_tree', 'kd_tree', 'brute'),
    'knr__weights': ('uniform', 'distance')

}

grid = GridSearchCV(estimator=ens, param_grid=params, cv=2)

# grid.fit(x,y)             # UNCOMMENT FROM THIS LINE TO ACTUALLY RUN GRIDSEARCH

# print (grid.best_params_)

# print("\n\n===================================  For all state's data    ===================================\n\n")
# print("Mean Squared Error of Ensemble Regressor:", mse)
# print("R-squared Score:", ens.score(x_test_scaled, y_test))

```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error


x = data_daily.drop(["Dates", "mp"], axis=1)
y = data_daily["mp"]        #All states' data

print(x.head())

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler_a = StandardScaler()
x_train_scaled = scaler_a.fit_transform(x_train)
x_test_scaled = scaler_a.transform(x_test)          #Scaling test set and training set


rf = RandomForestRegressor(criterion='absolute_error', random_state=42)         #Declaring base regressors with optimum hyperparameters
gbr = GradientBoostingRegressor(criterion='friedman_mse', loss='squared_error', random_state=42)
knr = KNeighborsRegressor(algorithm='ball_tree', n_neighbors=5, weights='uniform')
lr = LinearRegression()             

ens_a = VotingRegressor(estimators=[
    ('rf', rf),
    ('gbr', gbr),
    ('knr', knr),
    ('lr', lr)], verbose=False)

ens_a.fit(x_train_scaled, y_train)
y_pred = ens_a.predict(x_test_scaled)
mse = mean_squared_error(y_test, y_pred)

plt.scatter(y_test, y_pred)                     #Plot to visualize the performance of the model
plt.xlabel('Actual Rainfall (in mm)')
plt.ylabel('Predicted Rainfall (in mm)')
plt.title('Model trained on all states\' data')
plt.show()


print("\n\n===================================  For distant state's [tn, kl, assam, jk] data    ===================================\n")
print("Mean Squared Error of Ensemble Regressor:", mse)
print("R-squared Score:", ens_a.score(x_test_scaled, y_test))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error



x = data_daily[["chat", "guj", "mh", "raja", "up"]]
y = data_daily["mp"]            #Neighbouring states' data


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


scaler_n = StandardScaler()
x_train_scaled = scaler_n.fit_transform(x_train)
x_test_scaled = scaler_n.transform(x_test)

rf = RandomForestRegressor(criterion='poisson', random_state=42)            #Base regressors with optimum hyperparameters
gbr = GradientBoostingRegressor(criterion='squared_error', loss='huber', random_state=42)
knr = KNeighborsRegressor(algorithm='auto', n_neighbors=9, p=1)
lr = LinearRegression()

ens_n = VotingRegressor(estimators=[
    ('rf', rf),
    ('gbr', gbr),
    ('knr', knr),
    ('lr', lr)], verbose=False)


ens_n.fit(x_train_scaled, y_train)

y_pred = ens_n.predict(x_test_scaled)
mse = mean_squared_error(y_test, y_pred)

plt.scatter(y_test, y_pred)             #Plot to visualize the performance of the model
plt.xlabel('Actual Rainfall (in mm)')
plt.ylabel('Predicted Rainfall (in mm)')
plt.title('Model trained on neighbouring states\' data')
plt.show()

print("\n\n===================================  For neighbouring state's [chat, guj, mh, raja, up] data    ===================================\n")
print("Mean Squared Error of Ensemble Regressor:", mse)
print("R-squared Score:", ens_n.score(x_test_scaled, y_test))

```
