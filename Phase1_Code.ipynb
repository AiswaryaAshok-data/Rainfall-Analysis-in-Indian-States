```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("chat_rainfall.csv", header=2, index_col="Dates")
df2 = pd.read_csv("mp_rainfall.csv", header=2, index_col="Dates")

df1 = df1.drop(columns="NORMAL (mm) ")
df2 = df2.drop(columns="NORMAL (mm) ")

print(df1.shape, df2.shape)

rainfall_df = pd.merge(df1, df2, left_index=True, right_index=True)
missing = rainfall_df.isnull().any()        #To check for missing values
print("Presence of Missing Values?\n", missing, "\n\n")     

rainfall_df = rainfall_df.rename(columns={"ACTUAL (mm) _x": "Chattisgarh", "ACTUAL (mm) _y": "MP"})

print(rainfall_df.head())
print(rainfall_df.tail())
print(rainfall_df.shape)
```

```python
df_2020 = rainfall_df.iloc[2191:2539]       # Selecting rows corresponding to the year 2020
print(df_2020.shape)


print(df_2020.head())
print(df_2020.tail())

corr20 = df_2020["Chattisgarh"].corr(df_2020["MP"])

print("\n", "The correlation between rainfall in Chattisgarh and Madhya Pradesh in the year 2020 was found to be:", corr20)


plt.show()
```

```python
df_2021 = rainfall_df.iloc[2539:2893]       # Selecting rows corresponding to the year 2021
print(df_2021.shape)


print(df_2021.head())
print(df_2021.tail())

corr21 = df_2021["Chattisgarh"].corr(df_2021["MP"])

print("\n", "The correlation between rainfall in Chattisgarh and Madhya Pradesh in the year 2021 was found to be:", corr21)


plt.figure(figsize=(20, 6))
plt.plot(df_2021["Chattisgarh"], label="Chattisgarh", color="red")
plt.plot(df_2021["MP"], label="MP", color="blue")
plt.title("Line Plot of Rainfall in Chattisgarh (red) and MP (blue) in 2021")
plt.xlabel("Date")
plt.ylabel("Rainfall (in mm)")
plt.xticks(rotation = 90, fontsize=3)
plt.legend()
plt.show()
```

```python
df_2022 = rainfall_df.iloc[2893:3252]       # Selecting rows corresponding to the year 2022
print(df_2022.shape)


print(df_2022.head())
print(df_2022.tail())

corr22 = df_2022["Chattisgarh"].corr(df_2022["MP"])

print("\n", "The correlation between rainfall in Chattisgarh and Madhya Pradesh in the year 2022 was found to be:", corr22)

plt.figure(figsize=(20, 6))
plt.plot(df_2022["Chattisgarh"], label="Chattisgarh", color="red")
plt.plot(df_2022["MP"], label="MP", color="blue")
plt.title("Line Plot of Rainfall in Chattisgarh (red) and MP (blue) in 2022")
plt.xlabel("Date")
plt.ylabel("Rainfall (in mm)")
plt.xticks(rotation = 90, fontsize=3)
plt.legend()
plt.show()
```

```python
df_2023 = rainfall_df.iloc[3252:3616]       # Selecting rows corresponding to the year 2023
print(df_2023.shape)


print(df_2023.head())
print(df_2023.tail())

corr23 = df_2023["Chattisgarh"].corr(df_2023["MP"])

print("\n", "The correlation between rainfall in Chattisgarh and Madhya Pradesh in the year 2023 was found to be:", corr23)



plt.figure(figsize=(20, 6))
plt.plot(df_2023["Chattisgarh"], label="Chattisgarh", color="red")
plt.plot(df_2023["MP"], label="MP", color="blue")
plt.title("Line Plot of Rainfall in Chattisgarh (red) and MP (blue) in 2023")
plt.xlabel("Date")
plt.ylabel("Rainfall (in mm)")
plt.xticks(rotation = 90, fontsize=3)
plt.legend()
plt.show()
```
